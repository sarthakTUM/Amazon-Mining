### Splitting SalesRank Column

```python
def get_cat(l):
    [(k, v)] = l.items()
    return k
def get_rank(l):
    [(k, v)] = l.items()
    return v
def replace_amp(l):
    if not type(l) == float:
        return l.replace("&amp;", "&")
    else:
        return l
```

```python
rankDf = DF63K[DF63K['salesRank'].notnull()].copy()
rankDf = rankDf[rankDf.salesRank != {}].reset_index()
```

### Getting Nouns and ProNouns

```python
from textblob import TextBlob
```

```python
def get_nouns(l):
    if not type(l) == float:
        blob = TextBlob(l)
        outSet = set()
        for wrd,form in blob.tags:
            if form == 'NN' or form =='NNS' or form == 'NNP' or form =='NNPS':
                outSet.add(wrd)
        return outSet
    else:
        outSet = set()
        return outSet
def get_proper_nouns(l):
    if not type(l) == float:
        blob = TextBlob(l)
        outSet = set()
        for wrd,form in blob.tags:
            if form == 'PRP' or form =='PRP$' or form == 'WP' or form =='WP$' :
                outSet.add(wrd)
        return outSet
    else:
        outSet = set()
        return outSet
```

```python
ReviewedProductsDF['title_Nouns'] = ReviewedProductsDF.title.apply(get_nouns)
ReviewedProductsDF['title_ProNouns'] = ReviewedProductsDF.title.apply(get_proper_nouns)
```
