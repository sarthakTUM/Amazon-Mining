```python
def preprocess(vectDF,filestr):
    tempDF = vectDF.copy()
    del tempDF['asin']
    del tempDF['categoryLvl2_x']
    del tempDF['id_asin']
    return tempDF
```

```python
def computeJS(tempDF,filename):
    from sklearn.metrics.pairwise import pairwise_distances
    jac_sim = 1 - pairwise_distances(tempDF, metric = "hamming")
    # optionally convert it to a DataFrame
    jac_sim = pd.DataFrame(jac_sim, index=tempDF.T.columns, columns=tempDF.T.columns)
    
    del tempDF
    gc.collect()

    jac_sim = jac_sim.where(np.triu(np.ones(jac_sim.shape)).astype(np.bool))
    jac_sim=jac_sim.stack()
    jac_sim.columns = ['Pa','Pb','Score']
      

    jac1 = jac_sim.copy()
    jac1 = jac1.reset_index()
    jac1.columns=['Pa','Pb','Score']
    jac1['Score'] = jac1.Score.round(2)
   
    jac1 = jac1[jac1.Pa != jac1.Pb]
    jac1 = jac1[jac1.Score != 0.73]
    #jac1 = jac1[jac1.Score != 0.87]
      

    return jac1
```

```python
from PIL import Image
import urllib.request
```

```python
def generate_image(URL, asin):
    with urllib.request.urlopen(URL) as url:
        file_name = asin+'jpg'
        with open(file_name, 'wb') as f:
            f.write(url.read())

    img = Image.open(file_name)

    img.show()
```
